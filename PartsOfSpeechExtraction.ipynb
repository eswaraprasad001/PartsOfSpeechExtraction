```python
for entity in sen.ents:
    print(entity.text + ' - ' + entity.label_ + ' - ' + str(spacy.explain(entity.label_)))
```

```python
for word in sen:
    print(f'{word.text:{12}} {word.pos_:{10}} {word.tag_:{8}} {spacy.explain(word.tag_)}')
```

# Main Program


```python
# This Block of code will read a text file and execute the given task 

import spacy
import re
import glob

sp = spacy.load('en_core_web_sm')
path="/content/alp.txt"

for file in glob.glob(path):
    with open(file, encoding='utf-8', errors='ignore') as file_in:
        text = file_in.read()
        
        words_pattern = '[a-zA-Z]+'
        lines = re.findall(words_pattern, text, flags=re.IGNORECASE)
        for line in lines:
            line = sp(line)
            for token in line:
               print(f'{token.text:{12}} {token.pos_:{10}} {token.tag_:{8}} {spacy.explain(token.tag_)}')
```
